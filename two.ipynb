```python
import pandas as pd
from tabulate import tabulate

# Load job data from CSV file into DataFrame
data_frame = pd.read_csv(r'C:\Users\HP\Desktop\AI_20_KDG\employee_data.csv')

def filter_jobs_by_experience(experience):
    experience = int(experience)
    return data_frame[data_frame['Experience'] == experience]

def filter_jobs_by_job_title(location_filtered_data, job_title):
    return location_filtered_data[location_filtered_data['JobTitle'] == job_title]

def sort_employee_ratings(location_filtered_data, job_title_filtered_data):
    sorted_data = job_title_filtered_data.sort_values(by=['Experience', 'JobTitle', 'Current Employee Rating'], ascending=[True, True, False])
    return sorted_data[['FirstName','LastName', 'EmpID', 'ADEmail']]

while True:
    print("Select an option:")
    print("1. Filter by experience")
    print("2. Filter by job title (after filtering by experience)")
    print("3. Sort employee ratings by experience and job title")
    print("4. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == '4':
        print("Exiting...")
        break
    elif choice == '1':
        experience = input("Enter the Experience: ")
        location_filtered_data = filter_jobs_by_experience(experience)
        print(tabulate(location_filtered_data[['FirstName','LastName', 'EmpID', 'ADEmail']], headers='keys', tablefmt='pretty'))
    elif choice == '2':
        if 'location_filtered_data' not in locals():
            print("Please filter by experience first.")
            continue
        job_title = input("Enter the job title: ")
        job_title_filtered_data = filter_jobs_by_job_title(location_filtered_data, job_title)
        print(tabulate(job_title_filtered_data[['FirstName','LastName', 'EmpID', 'ADEmail']], headers='keys', tablefmt='pretty'))
        
    else:
        print("Invalid option!")
```
