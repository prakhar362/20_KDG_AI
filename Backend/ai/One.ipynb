```python
import pandas as pd
from tabulate import tabulate

# Load job data from CSV file into DataFrame
data_frame = pd.read_csv(r'C:\Users\HP\Desktop\AI_20_KDG\job.csv')

def filter_jobs_by_location(location):
    return data_frame[data_frame['location'] == location]

def filter_jobs_by_job_title(location_filtered_data, job_title):
    return location_filtered_data[location_filtered_data['job_title'] == job_title]

while True:
    print("Select an option:")
    print("1. Filter by location")
    print("2. Filter by job title (after filtering by location)")
    print("3. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == '3':
        print("Exiting...")
        break
    elif choice == '1':
        location = input("Enter the location: ")
        location_filtered_data = filter_jobs_by_location(location)
        print(tabulate(location_filtered_data[['company_name', 'job_title', 'location', 'ctc', 'experience', 'posted']], headers='keys', tablefmt='pretty'))
    elif choice == '2':
        if 'location_filtered_data' not in locals():
            print("Please filter by location first.")
            continue
        job_title = input("Enter the job title: ")
        job_title_filtered_data = filter_jobs_by_job_title(location_filtered_data, job_title)
        print(tabulate(job_title_filtered_data[['company_name', 'job_title', 'location', 'ctc', 'experience', 'posted']], headers='keys', tablefmt='pretty'))
    else:
        print("Invalid option!")
```
